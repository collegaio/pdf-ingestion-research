```python
# load dataframes
import polars as pl

df_1 = pl.read_parquet("./datasets/cds/parquet/cu-boulder.parquet")
df_2 = pl.read_parquet("./datasets/cds/parquet/purdue.parquet")

print(df_1)
print(df_2)
```

```python
# determine schema
cols = df_1.columns

dtype_counts = {col: {} for col in cols}

for i, dtype in enumerate(df_1.dtypes):
    col = cols[i]

    if dtype not in dtype_counts[col]:
        dtype_counts[col][dtype] = 1
    else:
        dtype_counts[col][dtype] += 1

for i, dtype in enumerate(df_2.dtypes):
    col = cols[i]

    if dtype not in dtype_counts[col]:
        dtype_counts[col][dtype] = 1
    else:
        dtype_counts[col][dtype] += 1

# print(dtype_counts)

# combine datarames with possible null columns
## all columns should be null optional
## all numeric columns with different column types (int vs float) should be float64
## 
for col in dtype_counts:
    dtypes = [dtype for dtype in dtype_counts[col]]
    if pl.Null not in dtypes and len(dtypes) > 1:
        print(col, ":", dtypes)
        print(df_1[col])
        print(df_2[col])
        
```
